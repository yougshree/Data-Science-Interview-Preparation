```python
import matplotlib.pyplot as plt
#pyplot is a collection of function that make matplotlib work as MATLAB
```

```python
#create first chart
#dummy
x= [1,2,3,4,5]
y= [10,20, 25, 35, 45]
```

```python
#line chart
plt.plot(x,y)
plt.show()
```

```python
plt.figure(figsize=(6,4)) #cutomize figure size

#customize chart
plt.plot(x,y, color= 'green', marker='o', linestyle='dashed', linewidth=2,
         markersize= 12)
plt.title('Line Chart')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
x= [1,2,3,4,5]
y1= [10,20, 25, 35, 45]
y2= [20,30, 35, 45, 55]
```

```python
#multiple line
plt.plot(x,y1, label='Sales 2024')
plt.plot(x,y2, label= 'Sales 2025')
plt.title('year line char')
plt.xlabel('months')
plt.ylabel('sales')
plt.legend()
plt.show()
```

```python
#bar chart
x= ['A', 'B', 'C', 'D', 'E']
y= [10,20, 25, 35, 45]
```

```python
#bar chart: catergorical data
plt.bar(x,y)
plt.title('Bar Chart')
plt.show()
```

```python
import random
data = [random.randint(1, 10) for _ in range(100)]
```

```python
#histogram continious data
plt.hist(data, bins=10) #bins: how many parts in data
plt.title('Histogram')
plt.show()
```

```python
categories= ['A', 'B', 'C', 'D', 'E']
sales= [10,20, 25, 35, 45]
```

```python
#pie chart: whole data relationships
plt.pie(sales, labels=categories, autopct='%1.1f%%', )
plt.title('Pie Chart')
plt.show()
```

```python
plt.pie(sales, labels=categories, autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart')
plt.show()
```

```python
y1= [10,20, 25, 35, 45]
y2= [20,30, 35, 45, 55]
```

```python
#Scatter plot: relation between variables
plt.scatter(y1, y2)
plt.title('Scatter Plot')
plt.show()
```

```python
#subplot: used to show multiple charts in one figure


#data1
categories= ['Mon', 'Tue', 'Wed', 'Thu','Fri']
sales= [10,20, 25, 35, 45]

#data2
y1= [10,20, 25, 35, 45]
y2= [20,30, 35, 45, 55]

#fix size
plt.figure(figsize=(10,10))
```

```python
#subplot: used to show multiple charts in one figure
#1st plot: bar chart
plt.subplot(1,2,1) #row, col, pos
plt.bar(categories, sales)
plt.title('Daily sales Bar chart')
plt.xlabel('Week days')
plt.ylabel('Sales')

#2nd plot:scatter plot()
plt.subplot(1,2,2)
plt.scatter(categories, sales)
plt.title('Daily Sales Scatter plot')
plt.xlabel('Week days')
plt.ylabel('Sales')

plt.show()




```

```python
import pandas as pd
#data
data={
    'Month': ['Jan','Feb','Mar','Apr','May'],
    'Sales': [100,200,300,400,500]
}
df= pd.DataFrame(data)
df
```

```python
#bar
#plt.bar(x,y)
#plt.figure(figsize=(4,4 ))
plt.bar(df['Month'], df['Sales'], color= 'green')
plt.title('Monthly sales bar chart')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.show()
```

```python
#Save charts/plot

plt.bar(df['Month'], df['Sales'], color= 'green')
plt.title('Monthly sales bar chart')
plt.xlabel('Months')
plt.ylabel('Sales')
#Save charts/plot
plt.savefig("Monthly_Sales_from_df.png")

plt.show()
```
